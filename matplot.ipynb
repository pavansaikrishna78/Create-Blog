Importing Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Line Plot

```python
x=[2,5,6,7,8]
y=[3,7,4,12,5]
```

Scatter Plot

```python
plt.scatter(x,y)
plt.title("Line Graph")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line Graph")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show
```

Vertical Bar Graph

```python
x = [2,5,6,7,8]
y = [3,7,4,12,5]
x2 = [4,5,8,9,8]
y2 = [3,5,7,12,6]
```

```python
plt.bar(x,y,color='g',linewidth=6)
```

```python
plt.bar(x2,y2,color='r',linewidth=6)
```

```python
plt.barh(x,y,color='r',linewidth=6)
```

```python
plt.barh(x2,y2,color='b',linewidth=6)
```

```python
plt.barh(x2,y2,color='r',linewidth=6)
```

Bubble Graph

```python
plt.scatter(x,y,s=np.random.rand(5)*1000,c=np.random.randn(5))
plt.title("Line graph")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show()
```

Pie Chart

```python
plt.pie(x)
```

```python
data=[3,7,4,12,5]
ex = (0,0.1,0,0,0)
lb = ['a','b','c','d','e']
plt.pie(data,labels=lb,explode=ex,shadow=True)
plt.show
```
